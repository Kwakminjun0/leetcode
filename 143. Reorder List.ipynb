```python
class Solution(object):
    def reorderList(self, head):
        """
        :type head: ListNode
        :rtype: None Do not return anything, modify head in-place instead.
        """


        # Brute Force------------>>>>>>>>>>>


        # if( not head):
        #     return None

        # arr=[]
        # while head:
        #     arr.append(head.val)
        #     head=head.next

        # r=len(arr)-1
        # l=0
        # dummy = ListNode(0)
        # curr = dummy
        # while l<=r:
        #     curr.next=ListNode(arr[l])
        #     curr=curr.next
        #     if(l!=r):
        #         curr.next=ListNode(arr[r])
        #         curr=curr.next
        #     l+=1
        #     r-=1
        # return dummy.next



        # Optimised Approch---------->>>>>>>>


        if not head or not head.next:
            return
        
        # 1. We will find mid in our Linked List
        slow=head
        fast=head
        while fast and fast.next:
            fast=fast.next.next
            slow=slow.next
        

        # 2: Reverse the second half of the linked list
        prev = None
        curr = slow
        while curr:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        
        # 3: Merge the two halves of the linked list
        first = head
        second = prev
        while second.next:
            temp1 = first.next
            temp2 = second.next

            first.next = second
            second.next = temp1

            first = temp1
            second = temp2
        
```
