```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        stack1 = []
        stack2 = []

        # Push all nodes of l1 to stack1
        while l1:
            stack1.append(l1.val)
            l1 = l1.next

        # Push all nodes of l2 to stack2
        while l2:
            stack2.append(l2.val)
            l2 = l2.next

        carry = 0
        head = None

        # While there are elements in stacks or there is a carry
        while stack1 or stack2 or carry:
            sum = carry
            if stack1:
                sum += stack1.pop()
            if stack2:
                sum += stack2.pop()

            carry = sum // 10
            node = ListNode(sum % 10)
            node.next = head
            head = node

        return head

# Example usage
def list_to_linkedlist(lst):
    head = ListNode(lst[0])
    current = head
    for value in lst[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

def linkedlist_to_list(node):
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result
        
```
